```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('L:\\archive (1)\\Titanic.csv')
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
sns.heatmap(titanic.isnull(),yticklabels=False);
```

```python
# Convert 'age' column to numeric, handling errors with coerce to replace non-numeric values with NaN
titanic['age'] = pd.to_numeric(titanic['age'], errors='coerce')

# Group by 'pclass' and calculate the mean of 'age'
mean_age = titanic.groupby('pclass')['age'].mean()
print(mean_age)
```

```python
titanic.info()
```

```python
Sex = pd.get_dummies(titanic['sex'])
Sex.head()
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('age')
titanic['age'].plot.hist(edgecolor='k').autoscale(enable=True,axis='both',tight=True);
```

```python
figure = plt.figure(figsize=(10, 5))
plt.title('Titanic')
sns.countplot(x='sibsp', hue='pclass', data=titanic)
plt.xlabel('sibsp', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='pclass',y='age',data = titanic);
```
